```python
pop = pop.drop(["Estimates Base"], axis=1)
pop
```

```python
pop = pop.drop(["Population Estimate 2010", "Population Estimate 2011", "Population Estimate 2012", "Population Estimate 2013", "Population Estimate 2014", "Population Estimate 2015", "Population Estimate 2016", "Population Estimate 2017", "Population Estimate 2018"], axis=1)
pop
```

```python
pop = pop.drop(["Census"], axis=1)
pop
```

```python
pop = pop[['Geographic Area','Population Estimate 2019']][0:3194]
pop.tail(10)
```

```python
pop['County'] = pop['Geographic Area'].apply(lambda x: x.split(', ')[0][0:])
pop['County'] = pop['County'].apply(lambda x: x.split(' County')[0])
pop['County'] = pop['County'].apply(lambda x: x.split(' Parish')[0]) # for LA
pop['County'] = pop['County'].apply(lambda x: x.split(' City')[0]) # for VA
pop['State'] = pop['Geographic Area'].apply(lambda x: x.split(', ')[1])
#pop['State'] = pop['State'].apply(lambda x: x.split('State')[0])
pop.head(70)
```
